# Pollutant Rose

This notebook contains an example of how to create nice-looking pollutant rose figures.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.ticker import MultipleLocator
import seaborn as sns
from pathlib import Path

sns.set("notebook", "ticks", palette="colorblind", font_scale=1.25)
```

## Load the data
- Check the column names

```python
# read the csv (this is just a random **final** data file from any sensor on the website)

file_path = Path(r"/home/ekans/Documents/AirPartners/datasets/PM00218-PM1.csv")
df = pd.read_csv(file_path)
print(df.columns)
```

## Build the pollutant rose function

```python
# set the index
df.set_index("timestamp", inplace=True)

# only keep a few columns of interest
df = df[["WD", "WS", "NO"]]

# resample the data to be on a 5min time base
df = df.resample("5min").mean().dropna(how='any')

# # output the info
df.info()
```

```python
def pollutant_rose(data, wind_dirs, data_bins, data_col, wd_col, **kwargs):
    """
    Pollution Wind Rose plot adapted from Phil Hobson.
    
    https://gist.github.com/phobson/41b41bdd157a2bcf6e14
    """
    units = kwargs.pop("units", "ppb")
    palette = kwargs.pop("palette", None)
    wd_bins = kwargs.pop("wd_bins", np.arange(-7.5, 370, 15))
    fs = kwargs.pop("fs", 20)
    ax_pos = kwargs.pop("ax_pos", (1, 0.75))
    
    def _convert_dir(dirs):
        bar_dir = dirs*np.pi/180. - np.pi/dirs.shape[0]
        bar_width = 2*np.pi / dirs.shape[0]
        
        return bar_dir, bar_width
        
    def _labels(bins, units):
        labels = list()
        for l, r in zip(bins[:-1], bins[1:]):
            if l == bins[0]:
                labels.append("clean")
            elif np.isinf(r):
                labels.append(">{} {}".format(l, units))
            else:
                labels.append("{} - {} {}".format(l, r, units))
        
        return labels
    
    # build the list of data labels
    data_labels = _labels(data_bins, units)
    wd_labels = (wd_bins[:-1] + wd_bins[1:]) / 2
    
    
    # keep only the necessary data from the dataframe
    dframe = data[[data_col, wd_col]].dropna(how='any')
    dframe = dframe.query("{} <= 360".format(wd_col))
    
    # count the number of "clean" data points
    count_clean = dframe[dframe[data_col] < data_bins[2]].shape[0]
    
    rose_data = (
        dframe.assign(data_bin=lambda x: pd.cut(dframe[data_col], bins=data_bins, labels=data_labels, right=True))
                .assign(wd_bin=lambda x: pd.cut(dframe[wd_col], bins=wd_bins, labels=wd_labels, right=False))
                .replace({"wd_bin": {360: 0}})
                .groupby(by=["data_bin", "wd_bin"])
                .size()
                .unstack(level="data_bin")
                .fillna(0)
                .assign(clean=lambda x: count_clean / data.shape[0])
                .sort_index(axis=1)
                .applymap(lambda x: x / data.shape[0] * 100)
    )
    
    if palette is None:
        palette = sns.color_palette("inferno", n_colors=rose_data.shape[1])
        
    bar_dir, bar_width = _convert_dir(wind_dirs)
    
    # finally, build the figure
    fig = plt.gcf()
    
    ax = fig.add_subplot(111, polar=True)
    
    ax.set_theta_direction("clockwise")
    ax.set_theta_zero_location("N")
    
    for n, (c1, c2) in enumerate(zip(rose_data.columns[:-1], rose_data.columns[1:])):
        if n == 0:
            ax.bar(bar_dir, rose_data[c1].values, width=bar_width, 
                   color=palette[0], edgecolor="none", label=c1, lw=0)
        
        ax.bar(bar_dir, rose_data[c2].values, width=bar_width, 
               bottom=rose_data.cumsum(axis=1)[c1].values, color=palette[n+1], edgecolor="none",
              label=c2, linewidth=0)
        
    # add the legend
    leg = ax.legend(loc=ax_pos, ncol=1, fontsize=fs)
    
    # set the tick labels
    ax.set_xticklabels(["N", "NE", "E", "SE", "S", "SW", "W", "NW"], fontsize=fs)
    
    # remove the ytick labels
    ax.set_yticklabels([])
```

## Create NO Figure

```python
fig = plt.figure(figsize=(8, 8))
    
pollutant_rose(
    data=df, wind_dirs=np.arange(0, 360, 15), 
    data_bins=[-1, 0, 5, 10, 15, 20, 25, np.inf],
    data_col="NO", wd_col="WD", 
    palette=sns.color_palette("viridis", 8),
)

# add a title
fig.suptitle("NO in Boston", fontsize=28);
```
