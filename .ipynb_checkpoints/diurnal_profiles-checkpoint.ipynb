# Diurnal Figures

This notebook contains an example of how to build nice diurnal figures.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.ticker import MultipleLocator
import seaborn as sns
from pathlib import Path

sns.set("notebook", "ticks", palette="colorblind", font_scale=1.25)
```

## Load data

```python
# read the csv (this is just a random **final** data file from any sensor on the website)
# file_path = Path(r"/home/ekans/Documents/AirPartners/airquality-plots/datasets/PM00218-PM_04-08-22.csv")
file_path = "datasets/PM00218-PM_04-08-22.csv"
df = pd.read_csv(file_path)
print(df.columns)
```

```python
# convert the timestamp_local column to a datetime object
df["timestamp_local"] = df["timestamp_local"].map(pd.to_datetime)

# set the index
df.set_index("timestamp_local", inplace=True)

# only keep a few columns of interest
df = df[["co2", "co", "no", "no2", "o3", "pm1", "pm25", "pm10"]]

# resample the data to be on a 5min time base
df = df.resample("5min").mean().dropna(how='all')

# output the info
df.info()
```

## Compute the Diurnal Statistics

```python
stats = df.groupby([df.index.hour, df.index.minute], as_index=False).describe()

stats.info()
```

## Build the Figure

Here is an example of CO2

```python
fig, ax = plt.subplots(1, figsize=(10, 6))

# get the index
idx = stats.index.values

# create an artificial time series where the start date is arbitrary
idx = pd.date_range(start="2020-01-01", periods=len(idx), freq='5min')

# choose the pollutant
col = "co2"

# plot the mean
ax.plot(idx, stats[col]["mean"], lw=3)

# plot the IQR as a shaded box
ax.fill_between(idx, y1=stats[col]["25%"], y2=stats[col]["75%"], alpha=0.25, lw=2)

# fix the tick marks
ticks = ax.get_xticks()

ax.set_xticks(np.linspace(ticks[0], ticks[-1], 5))
ax.set_xlim(ticks[0], ticks[-1])

# format the xticks
ax.xaxis.set_major_formatter(mdates.DateFormatter("%I:%M\n%p"))
ax.yaxis.set_minor_locator(MultipleLocator(100))

# label the Y axis
ax.set_ylabel("{}".format(col.upper()))

sns.despine()
plt.tight_layout()
```
